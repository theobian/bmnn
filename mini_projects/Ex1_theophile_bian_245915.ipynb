**Exponential Integrate and Fire Model:**

_Formula:_

$\tau{{\text{d}}\over{\text{d}}t}u=-(u-u_{\rm rest})+\Delta_{T}\\exp\left({u-%
\vartheta_{rh}\over\Delta_{T}}\right)+RI\$

We are going to use the Exponential I&F and adapt it to transform it into quadratic I&F.

**Quadratic Integrate and Fire Model:**

_Formula:_

$\tau{{\text{d}}\over{\text{d}}t}u=a_{0}\(u-u_{\rm rest})\(u-u_{c})+RI\$

_Parameters:_
- A0 > 0 
- uc > urest
- For I = 0 and initial condition u  <uc, the voltage decays to the resting potential urest.
- For u  >uc it increases so that an action potential is triggered.
- For numerical implementations of the model, the integratio is stopped if the voltage reaches a numerical threshold θreset and restarted with a reset value ur as new initial condition.
- For a mathematical analysis of the model, however, the standard assumption is θreset→∞ and ur→−∞.

- Defining some plotting functions because the functions from neurodynex plot specific things.

```python
def plot_voltage(voltage_monitor, spike_monitor, I_stim, plot_spikes=True, plot_current=True, title = "Voltage vs. Time"):
    fig, ax = plt.subplots(figsize = (16,4)) 
    ax.grid(linewidth=0.25)
    ax.plot(voltage_monitor.t/b2.ms, voltage_monitor.v[0]/b2.mV, color = "blue",label = "Voltage")
    
    if plot_spikes:
        for l in range(len(spike_monitor.t)):
            if (l==0):
                ax.axvline(spike_monitor.t[l]/b2.ms, ls='--', c = 'red', lw=1, label = "Spikes")
            ax.axvline(spike_monitor.t[l]/b2.ms, ls='--', c = 'red', lw=1)
            
    if plot_current:
        ax2 = ax.twinx()
        ax2.plot(I_stim.values/b2.nA, "--", color = "grey", label = "Input Current")
        ax2.set_ylabel("Current (nAmp)")
        lines, labels = ax.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax.legend(lines + lines2, labels + labels2, loc="best")
           
    ax.set_xlabel("Time (ms)")
    ax.set_ylabel("Voltage (mV)")
    ax.set_title(title)
    
    if not plot_current:
        ax.legend(loc="best")
        
    return fig, ax
```

- redefining the simulate_exponential_IF_neuron function from neurodynex into one that models a quadratic integrate and fire neuron.
     - for now we are using values taken from the exponential model and otherwise values that are not "optimized" for this model yet. We will modify these later on.

```python
### QUADRATIC
# take parameters from exponential IF neuron module in neurodynex3
# for new parameters, for now just guess, we will compute them later


# parameters
b2.defaultclock.dt = 0.05 * b2.ms

# a technical threshold to tell the algorithm when to reset vm to v_reset
FIRING_THRESHOLD_v_spike = -30. * b2.mV


# default values
MEMBRANE_TIME_SCALE_tau = 12.0 * b2.ms
MEMBRANE_RESISTANCE_R = 20.0 * b2.Mohm

V_REST = -65.0 * b2.mV


V_RESET = -65.0 * b2.mV
V_C = -35.0* b2.mV
A0 = (1.0/b2.mV)


# function
def simulate_quadratic_IF_neuron(
        tau=MEMBRANE_TIME_SCALE_tau,
        R=MEMBRANE_RESISTANCE_R,
        v_rest=V_REST,
        v_reset=V_RESET,
        v_c = V_C,
        a0 = A0,
        v_spike=FIRING_THRESHOLD_v_spike,
        I_stim=input_factory.get_zero_current(),
        simulation_time=200 * b2.ms):
    """
    Implements the dynamics of the quadratic 
    Integrate-and-fire model

    Args:
        tau (Quantity): Membrane time constant
        R (Quantity): Membrane resistance
        v_rest (Quantity): Resting potential
        v_reset (Quantity): Reset value (vm after spike)
        v_c (Quantity):  critical voltage for spike initiation by a short current pulse.
        a0 (Quantity): nonlinearity scaling term
        v_spike (Quantity) : voltage threshold for the spike condition
        I_stim (TimedArray): Input current
        simulation_time (Quantity): Duration for which the model is simulated

    Returns:
        (voltage_monitor, spike_monitor):
        A b2.StateMonitor for the variable "v" and a b2.SpikeMonitor
    """

    eqs = """
    dv/dt= ( a0 * (v-v_rest)*(v-v_c) + R*I_stim(t, i) ) / tau  : volt
    """
    neuron = b2.NeuronGroup(1, model=eqs, reset="v=v_reset", threshold="v>v_spike", method="euler")
    neuron.v = v_rest
    # monitoring membrane potential of neuron and injecting current
    voltage_monitor = b2.StateMonitor(neuron, ["v"], record=True)
    spike_monitor = b2.SpikeMonitor(neuron)

    # run the simulation
    net = b2.Network(neuron, voltage_monitor, spike_monitor)
    net.run(simulation_time)

    return voltage_monitor, spike_monitor
```

    - The quadratic model was made to be equal to the exponential model at the _rheobase_threshold_ and those two models are not exactly the same overall. When we move away from this voltage, which is of -55 mV (the inflexion point of the curve of each spike event), the two models behave differently.
    - the exponential model is linear in the region of the rheobase current
        - before the rheobase voltage, the quadratic model has a higher slope in that specific region.
    - eventually we reach the threshold and we move away from it. The nonlinearity starts to behave exponentially and the slope from the exponential models catches up to the one in the quadratic model.
        - after the rheobase voltage, the exponential model eventually reaches a higher slope (towards infinity)

Other observations:    
- the decrease to v_reset is sharp in both and follows the same logic as above

_Formula:_

 \\[.2cm]
\end{array}\right.\end{split}$

- some other plotting functions
- a function to get nullclines of the Exponential Integrate-and-Fire model (with a single adaptation current)


```python
# the plot_adex_state() plotting function used in the neurodynex library returns plots ...
# ... that are not customized as I would customize them so:
def plot_phase_plane(v_adex, title = ""):
    fig, ax = plt.subplots(figsize = (15,4), ncols = 3)


    ax[0].plot(v_adex.t/b2.ms, v_adex.w[0])
    ax[0].set_xlabel("t (ms)")
    ax[0].set_ylabel("w (A)")
    ax[0].set_title("W vs. time "+title, pad = 20)
    ax[0].grid()

    ax[1].plot(v_adex.t/b2.ms, v_adex.v[0])
    ax[1].set_xlabel("t (ms)")
    ax[1].set_ylabel("v (mV)")
    ax[1].set_title("V vs. time "+ title, pad = 20)
    ax[1].grid()

    ax[2].plot(v_adex.v[0], v_adex.w[0])
    ax[2].set_xlabel("v")
    ax[2].set_ylabel("w")
    ax[2].set_title("v vs. w phase plane; "+ title, pad = 20)
    ax[2].grid()

    fig.tight_layout()
    return fig, ax



def plot_phase_plane_nullclines(v_adex, v, vnullcline, wnullcline, plot_v = True, title = "nullclines"):
    fig, ax = plt.subplots()    
    ax.plot(v, v_nullcline, label = "v nullcline")
    ax.plot(v, w_nullcline, label = "w nullcline")
    
    if plot_v:
        ax.plot(v_adex.v[0], v_adex.w[0])
    ax.set_xlabel("v")
    ax.set_ylabel("w")
    ax.set_title(title, pad = 20)
    ax.legend(loc="best")
    ax.grid()
    
    return fig, ax


def get_nullclines(V_REST, SHARPNESS_delta_T, RHEOBASE_THRESHOLD_v_rh, MEMBRANE_RESISTANCE_R, I, ADAPTATION_VOLTAGE_COUPLING_a):
    # nullcline v:
    v = np.linspace(-40, -80, 1000)*b2.mV
    v_nullcline = (-v+V_REST + SHARPNESS_delta_T*np.exp((v-RHEOBASE_THRESHOLD_v_rh)/SHARPNESS_delta_T))/MEMBRANE_RESISTANCE_R + I

    # nullcline_w:
    w_nullcline = ADAPTATION_VOLTAGE_COUPLING_a*(v - V_REST)
    
    return v, v_nullcline, w_nullcline
```
