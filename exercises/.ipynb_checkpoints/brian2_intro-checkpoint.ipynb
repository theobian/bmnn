# NEURONS
###### Models
- all models are defined by systems of differential equations:
tau = 10*ms\
eqs = 
```
dv/dt=(1-v)/tau:1
```
- the :1 is :unit where unit is the SI unit of the variable in question - not the unit of the sides of the equation but of the variable in question (here: mu). But of course the equation needs to be dimensionally consistent for it to be usable by the pkg. It's a good idea to keep equations dimension-less to rid off any possible issues with ms vs ns vs s (e.g.)

###### Neuron Group simulation
G = NeuronGroup(1, eqs, method = "exact")\
- where 1 is the number of neurons and eqs its defining diffeq.
- use ```start_scope()``` to make sure brian2 objects are simulation-bounded. Basically make sure that Brian objects created before the function are not included in the next run of the simulation
- by default all variables start with value 0.
- method for solving diffeq; could also use "euler" or any other

###### Recording behavior: states, spikes
State = StateMonitor(G, "v", record = True)\
run(100*ms)\
plot(State.t/ms, State.v\[0\])\
- ```StateMonitor``` object is used to record values of a neuron variable WHILE the simulation runs. First argument is where to record form, second is which variable to record. If use record = 0 it means we record al values for neuron 0.
G = NeuronGroup(1, eqs, threshold = "v>0.8", reset = "v=0", method = "exact")\
Spike = Spikemonitor(G)\
plot(Spike.t/ms, Spike.v\[0\])\
- spiking behavior: ```SpikeMonitor``` object takes group of neurons whose spike you want to record as its argument and stores the spikes in the variable ```t```.

###### Refractoriness
```
dv/dt = (1-v)/tau : 1 (unless refractory)
```
G = NeuronGroup(..., refractory = 5*ms, ...)\
- this keyword means that the neuron can't spike in this period BUT it doesn't change how v behaves... To do that: use ```(unless refractory)``` at the end of the differential equation. This makes v stay constant during the refractory period!

###### Multiple Neurons
N = 100; tau = 10*ms; eqs = ```dv/dt = (2-v)\tau :1```;\

- N = number of neurons obvi

###### Parameters

eqs = 
```
dv/dt = (v0-v)/tau :1 (unless refractory)
v0 :1
```
G = ...; Spike = ...; duration = 1000*ms;\
G.v0 = "i*v0_max/(N-1)"\
run(duration)\
plot(G.v0, Spike.count/duration)\
- declare a per-neuron parameter v0 qith units 1 (dimensionless)
    - need to initiate parameter: G.v0 = ...
    - in that equation, ```i``` refers to neuron index
- plot: count variable of SpikeMonitor is an array of number of spikes each neuron in the group fired.

###### stochastic Neurons
- usually include randomness when modeling neurons (neural noise)
    - do this using: xi in differential equations. It's a stochastic differential, but can be thought of as a Gaussian random variable with mean 0 and standard deviation 1.
    - need to take into account the way stochastic differentials scale with time, so need to multiply it by (tau**-0.5) in the eqs.
    - need to use "euler" because "exact" won't work there.\
...\
eqs = 
```
dv/dt = (v_drive+vr-v)/tau + sigma*xi*tau**-0.5 : volt
dvt/dt = (vt0-vt)/tau_t : volt
```    
...\
hist(Spike.t/ms, 100, histtype="stepfilled", facecolor="k",weights = ones(len(Spike))/(N*defaultclock.dt)

````python
start_scope()

N = 100
tau = 10*ms
v0_max = 3.
duration = 1000*ms
sigma = 0.2

eqs = '''
dv/dt = (v0-v)/tau+sigma*xi*tau**-0.5 : 1 (unless refractory)
v0 : 1
'''

G = NeuronGroup(N, eqs, threshold='v>1', reset='v=0', refractory=5*ms, method='euler')
M = SpikeMonitor(G)

G.v0 = 'i*v0_max/(N-1)'

run(duration)

figure(figsize=(12,4))
subplot(121)
plot(M.t/ms, M.i, '.k')
xlabel('Time (ms)')
ylabel('Neuron index')
subplot(122)
plot(G.v0, M.count/duration)
xlabel('v0')
ylabel('Firing rate (sp/s)');
````

````python
start_scope()

N = 1000
tau = 10*ms
vr = -70*mV
vt0 = -50*mV
delta_vt0 = 5*mV
tau_t = 100*ms
sigma = 0.5*(vt0-vr)
v_drive = 2*(vt0-vr)
duration = 100*ms

eqs = '''
dv/dt = (v_drive+vr-v)/tau + sigma*xi*tau**-0.5 : volt
dvt/dt = (vt0-vt)/tau_t : volt
'''

reset = '''
v = vr
vt += delta_vt0
'''

G = NeuronGroup(N, eqs, threshold='v>vt', reset=reset, refractory=5*ms, method='euler')
spikemon = SpikeMonitor(G)

G.v = 'rand()*(vt0-vr)+vr'
G.vt = vt0

run(duration)

_ = hist(spikemon.t/ms, 100, histtype='stepfilled', facecolor='k', weights=ones(len(spikemon))/(N*defaultclock.dt))
xlabel('Time (ms)')
ylabel('Instantaneous firing rate (sp/s)');
````

# SYNAPSES
- connect neurons via synapses!

###### One-to-one synapse
- e.g.: one that causes an instantaneous change in a variable after a spike:

````python
start_scope()

eqs = '''
dv/dt = (I-v)/tau : 1
I : 1
tau : second
'''
G = NeuronGroup(2, eqs, threshold='v>1', reset='v = 0', method='exact')
G.I = [2, 0]
G.tau_ = [10, 100]*ms

S = Synapses(G, G, on_pre='v_post += 0.2')
S.connect(i=0, j=1)

M = StateMonitor(G, 'v', record=True)

run(100*ms)

fig, ax = plt.subplots()
plot(M.t/ms, M.v[0], label='Neuron 0')
plot(M.t/ms, M.v[1], label='Neuron 1')
xlabel('Time (ms)')
ylabel('v')
legend(bbox_to_anchor=(0.75, -0.2), ncol=2)
````

- created 2 neurons, with the same eqn but different values for parameters I and tau. I is input current, tau is period. So neuron 1 won't spike on its own; neuron 0 spikes repeatedly at a fairly high rate.
- ```Synapses(source,target)```, here both the source and the target is the group of neurons G. 
- ```"on_pre="v_post += 0.2``` means that when a spike occurs in the presynaptic neuron (on pre) it causes a post change (instantaneous) which increases post by 0.2. Basically the spike increases V in post neuron by 0.2.
- THEN finally need to instantiate a synapses: S.connect(i=0,j=1) creates a synapse FROM 0 TO 1.

###### Adding a weight

````python
start_scope()

eqs = '''
dv/dt = (I-v)/tau : 1
I : 1
tau : second
'''
G = NeuronGroup(3, eqs, threshold='v>1', reset='v = 0', method='exact')
G.I = [2, 0, 0]
G.tau_ = [10, 100, 100]*ms

S = Synapses(G, G, 'w : 1', on_pre='v_post += w')
S.connect(i=0, j=[1,2])
S.w = 'j*0.2'

M = StateMonitor(G, 'v', record=True)

run(100*ms)

fig, ax = plt.subplots()
plot(M.t/ms, M.v[0], label='Neuron 0')
plot(M.t/ms, M.v[1], label='Neuron 1')
plot(M.t/ms, M.v[2], label='Neuron 2')
xlabel('Time (ms)')
ylabel('v')
legend(bbox_to_anchor=(0.95, -0.2), ncol=3)
````

- S.w = 'j*0.2' with j is the source neuron index

##### Spike-timing depending plasticity
- STDP function: The change in synaptic weight w is the sum over all presynaptic spike times t_pre and postsynaptic spike times t_post of some function W of the difference in these spike times \
$\Delta w = \sum_{t_{pre}} \sum_{t_{post}} W(t_{post}-t_{pre})$\
$\begin{split}W(\Delta t) = \begin{cases}
A_{pre} e^{-\Delta t/\tau_{pre}} & \Delta t>0 \\
A_{post} e^{\Delta t/\tau_{post}} & \Delta t<0
\end{cases}\end{split}$

Here is the function:

````python
tau_pre = tau_post = 20*ms
A_pre = 0.01
A_post = -A_pre*1.05
delta_t = linspace(-50, 50, 100)*ms
W = where(delta_t>0, A_pre*exp(-delta_t/tau_pre), A_post*exp(delta_t/tau_post))
plot(delta_t/ms, W)
xlabel(r'$\Delta t$ (ms)')
ylabel('W')
axhline(0, ls='-', c='k');
````

- if you use this: have to sum over all pairs of spikes. Not efficient, plus a neuron can't remember all its previous spike times...
- so: use a_pre and a_post, traces of synaptic activity s.t.:\
$\begin{split}\begin{align}
\tau_{pre}\frac{\mathrm{d}}{\mathrm{d}t} a_{pre} &= -a_{pre}\\
\tau_{post}\frac{\mathrm{d}}{\mathrm{d}t} a_{post} &= -a_{post}
\end{align}\end{split}$
- update traces upon new spike event (post and pre synaptic spikes)\
$\begin{split}\begin{align}
a_{pre} &\rightarrow a_{pre}+A_{pre}\\
w &\rightarrow w+a_{post}\\[0.3em]
a_{post} &\rightarrow a_{post}+A_{post}\\
w &\rightarrow w+a_{pre}
\end{align}\end{split}$
- can define synapses using w, apre, apost. 
- can use the keyword ```(event-driven)``` in synapse, it means that although the variables evolve over time continuously, Brian updates them only at the time of an event! very useful for the update rules above :p, we only need to update apre and apost upon spiking, not continuously!
- on_pre argument can receive a set of equations to update various parameters. A usual thing is to clip weights so they don't fall out of range using clip(x, low, high)
- on_post is similar.
- ```(clock-driven)``` in synapse is continuous update (vs. event-driven)

````python
start_scope()

taupre = taupost = 20*ms
wmax= 0.01
Apre = 0.01
Apost = -Apre*taupre/taupost*1.05

weights_rule = \
'''
w:1
dapre/dt = -apre/taupre:1 (clock-driven)
dapost/dt = -apost/taupost:1 (clock-driven)
'''
on_pre_rule = \
'''
v_post += w
apre += Apre
w = clip(w+apost, 0, wmax)
'''
on_post_rule = \
'''
apost += Apost
w = clip(w+apre, 0, wmax)
'''

G = NeuronGroup(2, 'v:1', threshold='t>(1+i)*10*ms', refractory=100*ms)
# using the threshold rule above: neuron 0 fires at 10ms 
# while neuron 1 fires at 20ms
S = Synapses(G, G,weights_rule, on_pre = on_pre_rule, on_post = on_post_rule, method = linear)
S.connect(i=0, j=1)
M = StateMonitor(S, ['w', 'apre', 'apost'], record = True)

run(30*ms)

fig,ax  = plt.subplots(ncols=2, figsize = (10,5))
ax[0].plot(M.t/ms, M.apre[0], label='apre')
ax[0].plot(M.t/ms, M.apost[0], label = 'apost')
ax[1].plot(M.t/ms, M.w[0], label = 'w')
ax[0].set_xlabel('Time(ms)')
ax[0].set_ylabel('Voltage')
ax[1].set_xlabel('Time(ms)')
ax[0].legend(loc="best")
ax[1].legend(loc="best")
````

# SIMULATIONS

###### Multiple runs
 - changing parameters of one simulation 
- just use a for loop? if you recreate the objects everytime it's s-l-o-w!
--> set up the network once and copy state before loop, restore it at beginning of iteration!
    - use ```store()``` and ```restore()``` to store the current state of the network and restore the original/stored state of the network

````python
start_scope()

num_inputs = 100
input_rate = 10*Hz
weight = 0.1
tau_range = linspace(1, 10, 30)*ms
output_rates = []
eqs = \
'''
dv/dt = -v/tau :1
'''

P = PoissonGroup(num_inputs, rates = input_rate)
G = NeuronGroup(1,eqs, threshold = "v>1", reset = "v=0", method = "exact")
S = Synapses(P, G, on_pre = "v+=weight")
S.connect()
M = SpikeMonitor(G)

store()

for tau in tau_range:
    restore()
    run(1*second)
    output_rates.append(M.num_spikes/second)

plot(tau_range/ms, output_rates)
xlabel(r'$\tau$ (ms)')
ylabel('Firing rate (sp/s)')
````

- above: a lot of noise : because we refresh the Poisson group which has random distr.
- ```Network``` objects: allow us to choose which objects to be included in the simulation.

````python
start_scope()

num_inpus = 100
input_rate = 10*Hz
weight = 0.1
tau_range = linspace(1, 10, 30)*ms
output_rates = []
eqs = \
'''
dv/dt = -v/tau :1
'''
# construct a network with poisson group, run it ONCE and save spike times/indices
P = PoissonGroup(num_inputs, rates = input_rate)
M_p = SpikeMonitor(P)
net = Network(P, M_p)

net.run(1*second)
spikes_i = M_p.i
spikes_t = M_p.t

# construct a network with SGG, G, and run it a bunch of times; restore each time
SGG = SpikeGeneratorGroup(num_inputs, spikes_i, spikes_t)
G = NeuronGroup(1, eqs, threshold = "v>1", reset = "v=0", method = "exact")
S = Synapses(SGG, G, on_pre = "v+=weight")
S.connect()
M_g = SpikeMonitor(G)

net = Network(SGG, G, S, M_g)
net.store()

for tau in tau_range:
    net.restore()
    net.run(1*second)
    output_rates.append(M_g.num_spikes/second)

plot(tau_range/ms, output_rates)
xlabel(r'$\tau$ (ms)')
ylabel('Firing rate (sp/s)')
````

###### Changing things during a run

- e.g. changing the input current of a neuron
- to do that we can use the method ```run_regularly``` which lets you run code specific to a single NeuronGroup every dt, instead of having to loop and use ```run```.
- can also use ```network_operation```

Try with a Hodgkin-Huxley Neuron:

````python
start_scope()

area = 20000*umetre**2
Cm = 1*ufarad*cm**-2 * area
g1 = 5e-5*siemens*cm**-2 * area
E1 = -65*mV
EK = -90*mV
ENa = 50*mV
g_na = 100*msiemens*cm**-2 * area
g_kd = 30*msiemens*cm**-2 * area
VT = -63*mV
eqs_HH = \
'''
dv/dt = (gl*(El-v) - g_na*(m*m*m)*h*(v-ENa) - g_kd*(n*n*n*n)*(v-EK) + I)/Cm :volt
dm/dt = 0.32*(mV**-1)*(13.*mV-v+VT)/
    (exp((13.*mV-v+VT)/(4.*mV))-1.)/ms*(1-m)-0.28*(mV**-1)*(v-VT-40.*mV)/
    (exp((v-VT-40.*mV)/(5.*mV))-1.)/ms*m : 1
dn/dt = 0.032*(mV**-1)*(15.*mV-v+VT)/
    (exp((15.*mV-v+VT)/(5.*mV))-1.)/ms*(1.-n)-.5*exp((10.*mV-v+VT)/(40.*mV))/ms*n : 1
dh/dt = 0.128*exp((17.*mV-v+VT)/(18.*mV))/ms*(1.-h)-4./(1+exp((40.*mV-v+VT)/(5.*mV)))/ms*h : 1
I : amp
'''





N = NeuronGroup(1, 
                    eqs_HH,
                    threshold = "v>-40*mV",
                    refractory = "v>-40*mV",
                    method = "exponential_euler")
N.v = E1

M_state = StateMonitor(N, "v", record = True)
M_spike = SpikeMonitor(N, variables = "v")

N.run_regularly("I = rand()*50*nA", dt = 10*ms)

# IF using network operation INSTEAD:
# @network_operation(dt=10*ms)
# def change_I():
#     N.I = rand()*50*nA

run(50*ms)

figure(figsize=(15,7))
for l in range(5):
    axvline(l*10, ls='--', c = 'k', lw=3)
axhline(E1/mV, ls='-', c='lightgray', lw=3)
plot(M_state.t/ms, M_state.v[0]/mV, '-b')
plot(M_spike.t/ms, M_spike.v/mV, 'ob')
xlabel('Time (ms)')
ylabel('v (mV)')
````

- now let's have different neurons with different capacitances and synapses
- need to make sure that the variable ```i``` which should be the same for each neuron, and is initialized in a random manner at each run, IS actually the same --> add ```(shared)``` keyword in eqn initialization for this parameter. 
    - This is because "run_regularly" code, like "threshold" and "reset" statements, are naturally interpeted as being run separately for each neuron

````python
start_scope()

N = 3
eqs_HH_2 = '''
dv/dt = (gl*(El-v) - g_na*(m*m*m)*h*(v-ENa) - g_kd*(n*n*n*n)*(v-EK) + I)/C : volt
dm/dt = 0.32*(mV**-1)*(13.*mV-v+VT)/
    (exp((13.*mV-v+VT)/(4.*mV))-1.)/ms*(1-m)-0.28*(mV**-1)*(v-VT-40.*mV)/
    (exp((v-VT-40.*mV)/(5.*mV))-1.)/ms*m : 1
dn/dt = 0.032*(mV**-1)*(15.*mV-v+VT)/
    (exp((15.*mV-v+VT)/(5.*mV))-1.)/ms*(1.-n)-.5*exp((10.*mV-v+VT)/(40.*mV))/ms*n : 1
dh/dt = 0.128*exp((17.*mV-v+VT)/(18.*mV))/ms*(1.-h)-4./(1+exp((40.*mV-v+VT)/(5.*mV)))/ms*h : 1
I : amp (shared)
C : farad
'''

group = NeuronGroup(N, eqs_HH_2,
                    threshold='v > -40*mV',
                    refractory='v > -40*mV',
                    method='exponential_euler')
group.v = El

group.C = array([0.8, 1, 1.2])*ufarad*cm**-2*area

statemon = StateMonitor(group, variables=True, record=True)

group.run_regularly('I = rand()*50*nA', dt=10*ms)
run(50*ms)


figure(figsize=(9, 4))
for l in range(5):
    axvline(l*10, ls='--', c='k')
axhline(El/mV, ls='-', c='lightgray', lw=3)
plot(statemon.t/ms, statemon.v.T/mV, '-')
xlabel('Time (ms)')
ylabel('v (mV)');
````

###### Adding input
- just change I the input current to a function: e.g. a sine wave

````python
start_scope()

A = 2.5
f = 10*Hz
tau = 5*ms
eqs = \
'''
dv/dt = (I-v)/tau :1
I = A*sin(2*pi*f*t) :1
'''

G = NeuronGroup(1, eqs, threshold = "v>1", reset = "v=0", method = "euler")
M = StateMonitor(G, variables= True, record = True)

run(200*ms)

plot(M.t/ms, M.v[0], label = 'v')
plot(M.t/ms, M.I[0], label = 'I')
xlabel("Time (ms)")
ylabel("Voltage")
legend(loc="best")
````
